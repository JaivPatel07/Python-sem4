# Pandas Series 

A Pandas Series is a fundamental, one-dimensional labeled array data structure in the Pandas library for Python. It is capable of holding data of any type (integers, strings, floats, Python objects, etc.) and is similar to a single column in a spreadsheet or a table

```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(myvar)
```

```python
print(type(myvar))
```

```python
print(myvar[1])
```

```python
# for usermade index

a=[1,7,2]# i can take tuple also
myvar=pd.Series(a,index=['x','y','z'])
print(myvar)
print(type(myvar))
```

```python
# print(myvar[1]) this is depricated so no longer available
print(myvar['y'])
```

```python
a=(1.2,2,3,4,5,6)
myvar=pd.Series(a)
print(myvar)
print(type(myvar))
```

```python
student_name=['ram','shyam','seeta','geeta']
marks=[60,70,80,90]
myvar=pd.Series(marks,index=student_name)
print(myvar)
```
