# Pandas Series 

A Pandas Series is a fundamental, one-dimensional labeled array data structure in the Pandas library for Python. It is capable of holding data of any type (integers, strings, floats, Python objects, etc.) and is similar to a single column in a spreadsheet or a table

```python
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(myvar)
```

```python
print(type(myvar))
```

```python
print(myvar[1])
```

```python
# for usermade index

a=[1,7,2]# i can take tuple also
myvar=pd.Series(a,index=['x','y','z'])
print(myvar)
print(type(myvar))
```

```python
# print(myvar[1]) this is depricated so no longer available
print(myvar['y'])
```

```python
a=(1.2,2,3,4,5,6)
myvar=pd.Series(a)
print(myvar)
print(type(myvar))
```

```python
student_name=['ram','shyam','seeta','geeta']
marks=[60,70,80,90]
myvar=pd.Series(marks,index=student_name)
print(myvar)
```

# Data frame

A pandas DataFrame is a two-dimensional, size-mutable, tabular data structure in Python, much like a spreadsheet or SQL table, used for data analysis and manipulation. It is the primary data structure in the pandas library and is built on top of the NumPy package.

```python
import pandas as pd
data={'calories':[420,380,400,370],'duration':[50,40,45,35]}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.loc[0])
```

```python
print(df.iloc[0])#  a property in Python's Pandas library used for purely integer-location-based indexing
```

```python
print(df['calories'])
```

```python
print(df.loc[[0,1]])
```

```python
df=pd.DataFrame(data,index=['day1','day2','day3','day4'])
print(df)
```

```python
print(df.loc['day1'])
```

```python
print(df.iloc['day1']) # index in form of String so iloc not work
```

```python
print(df[::2])
```

## set_index()

The set_index() method in the pandas library is used to set one or more existing columns or arrays as the index (row labels) of a DataFrame. This is useful for efficient data retrieval, alignment, and reorganization, especially when working with meaningful identifiers like IDs or dates.

```python
import pandas as pd
data={'Name':['Alice','Bob','charlie'],
     'Age':[25,30,35],
     'city':['New York','Los Angeles','chicago']}
df=pd.DataFrame(data)
print(df)
```

```python
df=df.set_index('Name')
print(df)
```

```python
df.set_index('Name',drop=False)
```

```python
df.set_index('Name',drop=False,inplace=True)# inplace update main df
df
```

```python
a=df.set_index(['Name','Age'],drop=False)
a
```

## reset_index()

The reset_index() method in the pandas library for Python is used to reset the DataFrame's index to the default integer index (0, 1, 2, ...).

```python
df=pd.DataFrame(data,index=['a','b','c'])
df
```

```python
df_reset=df.reset_index()
df_reset
```

```python
df_reset=df.reset_index(drop=True)
df_reset
```
